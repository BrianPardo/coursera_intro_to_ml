In this exercise, we implement the backpropagation algorithm for neural networks and apply it to the task of hand-written digit recognition.

```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize
from numpy import matmul as mm

# Import data and load into x and y
my_data = scipy.io.loadmat('ex4data1.mat')
x = my_data['X']
y = my_data['y']

# Map 10 to 0 since python is 0-indexed
y[y==10] = 0

(m, n) = np.shape(x)
```

```python
def display_data(x):
    # m is number of samples, n is number of pixels
    (m, n) = np.shape(x)
    
    # number of rows/columns of pixels and images
    pad = 1
    l_pix = int(np.floor(np.sqrt(n)))
    l_img = int(np.ceil(np.sqrt(m)))
    w = l_pix*l_img+l_img*pad-1
    
    # initialize the grid
    img_grid = np.ones((w,w))
    
    for i, img in enumerate(x):
        # compute the start and end pixel for each image w/ pad
        h_start = (i%l_img)*l_pix+pad*(i%l_img)
        h_end = h_start+l_pix
        v_start = ((i)//l_img)*l_pix+pad*(i//l_img)
        v_end = v_start+l_pix
        
        # reshape image into square and replace image in grid
        image = img.reshape(l_pix, l_pix).T
        img_grid[v_start:v_end, h_start:h_end] = image
        
    return img_grid
```

```python
# Select a sample of the digits to display
sample = x[np.random.choice(m,100,replace=False)]

# We now display the images that have been
# reconstructed from the flat arrays
fig1, ax = plt.subplots(figsize=(5,5))
img_grid = display_data(sample)
ax.imshow(img_grid,cmap = 'bone_r')
ax.set_title('Sample of handwritten digits from training set')
ax.axis('off');
```

```python
# We now need to load the weights
my_weights = scipy.io.loadmat('ex4weights.mat')

theta1 = my_weights['Theta1']
theta2 = my_weights['Theta2']

# We need to shift the output layer because python is 0-indexed
theta2 = np.append(theta2[-1,:].reshape(1,-1),theta2[:-1,:], axis=0)

# Flatten theta1 and theta2
theta = np.append(theta1.flatten(), theta2.flatten())
```

```python
# Default parameters of the neural network
input_layer_size = 400
hidden_layer_size = 25
num_labels = 10
lambda_r = 1
```

```python
# It is also helpful to have a function to append a column of 1's to x
def add_ones(x):
    ones = np.ones((len(x),1))
    return np.append(ones, x, axis=1)

# We need the sigmoid function and the sigmoid gradient function
def sigmoid(z):
    return np.power(1 + np.exp(-z), -1)

def sigmoid_gradient(z):
    return sigmoid(z)*(1-sigmoid(z))
```

```python
def nn_cost_function(nn_params, 
                   input_layer_size, hidden_layer_size, num_labels,
                   x, y, lambda_r):
    
    m = len(x)
    # Reshape theta params
    theta1 = nn_params[:(input_layer_size+1)*(hidden_layer_size)].reshape(hidden_layer_size,
                                                                     input_layer_size+1)
    theta2 = nn_params[(input_layer_size+1)*(hidden_layer_size):].reshape(num_labels,
                                                                     hidden_layer_size+1)
    
    theta = (theta1, theta2)
    a = {0:0,1:0,2:0}
    z = {0:0,1:0}
    
    # Implement feed-forward algorithm
    a[0] = x
    for i, layer in enumerate(theta):
        # add ones column
        a[i] = add_ones(a[i])
        z[i] = mm(a[i],layer.T)
        a[i+1] = sigmoid(z[i])
    
    # Compute cost
    p = np.array([range(num_labels)]==y)
    
    cost = (1/m)*np.sum((0-p)*np.log(a[2])-(1-p)*np.log(1-a[2]));
    cost = cost + (lambda_r/(2*m))*(np.sum(theta1[:,1:]**2)+np.sum(theta2[:,1:]**2))
    
    # Compute deltas
    delta = {0:0,1:0}
    delta[1] = a[2] - p
    delta[0] = mm(delta[1],theta2[:, 1:])*sigmoid_gradient(z[0])
    
    # Compute gradients
    theta2_grad = (1/m)*mm(delta[1].T,a[1])
    theta1_grad = (1/m)*mm(delta[0].T,a[0])

    # Regularize gradients

    theta2_reg = (lambda_r/m)*theta2
    theta2_reg[:,0] = 0
    theta1_reg = (lambda_r/m)*theta1
    theta1_reg[:,0] = 0

    theta2_grad = theta2_grad + theta2_reg;
    theta1_grad = theta1_grad + theta1_reg;

    # Unroll gradients
    grad = np.append(theta1_grad.flatten(),theta2_grad.flatten())
    return float(cost), grad
```

```python
J, _ = nn_cost_function(theta, 400, 25, 10, x, y, lambda_r)
print(f'Cost at parameters (loaded from ex4weights): {J:.6f}.')
```

```python
# We next implement numerical gradient checking
def numeric_gradient(J, theta, opts):
    
    e = 1e-4
    numgrad = np.zeros(len(theta),)
    p = np.zeros(len(theta),)
    
    for i in range(len(theta)):
        p[i] = e
        Jp, _ = J(theta + p, *opts)
        Jm, _ = J(theta - p, *opts)  
        numgrad[i] = (Jp-Jm)/(2*e)
        p[i] = 0
    
    return numgrad
```

```python
def rand_initialize_weights(l_in, l_out):
    """Randomly initialize the weights of a layer with l_in
       incoming connections and l_out outgoing connections plus bias term"""

    epsilon_init = 0.12;
    w = np.random.rand(l_out, 1 + l_in)*2*epsilon_init - epsilon_init
    
    return w

def debug_initialize_weights(fan_out, fan_in):
    # Initialize w using "sin", this ensures that w is always of the same
    # values and will be useful for debugging
    num_el = fan_out*(1+fan_in)
    w = np.array(np.sin(range(1, 1 + num_el))/10)
    return w.reshape(fan_out, 1 + fan_in)
```

```python
def check_nn_gradients(lambda_r = 0):
    input_layer_size = 3;
    hidden_layer_size = 5;
    num_labels = 3;
    m = 5;

    # We generate some 'random' test data
    theta1 = debug_initialize_weights(hidden_layer_size, input_layer_size);
    theta2 = debug_initialize_weights(num_labels, hidden_layer_size);
    
    theta = np.append(theta1,theta2)
    
    x = debug_initialize_weights(m, input_layer_size - 1);
    y = np.array(range(1, m+1)).reshape(-1,1)
    y = 1 + (y % num_labels)
    
    opts = (input_layer_size, hidden_layer_size, num_labels, x, y, lambda_r)
    
    _, grad = nn_cost_function(theta, *opts)
    
    numgrad = numeric_gradient(nn_cost_function, theta, opts)
    print(f' Analytical |  Numerical\n   --------------------')
    for i in range(10):
        print(f'{grad[i]:11.5f} |{numgrad[i]:9.5f}')
    diff = np.linalg.norm(grad-numgrad)/np.linalg.norm(grad+numgrad)
    print(f'\nIf our backpropagation implementation is correct, then \nthe relative difference will be small (less than 1e-9). \n\nRelative Difference: {diff:.5} ')
```

```python
check_nn_gradients(1)
```

```python
# Randomly initialize theta
initial_theta1 = rand_initialize_weights(input_layer_size, hidden_layer_size)
initial_theta2 = rand_initialize_weights(hidden_layer_size, num_labels)
initial_theta = np.append(initial_theta1, initial_theta2)

inputs = (input_layer_size, hidden_layer_size, num_labels, x, y, lambda_r)
kwargs = {'jac':True, 'method':'TNC', 'options': {'maxiter':100}}

optim = minimize(nn_cost_function, initial_theta, inputs, **kwargs)

theta = optim.x

def predict(nn_params, input_layer_size, hidden_layer_size, num_labels, x):
    
    m = len(x)
    # Reshape theta params
    theta1 = nn_params[:(input_layer_size+1)*(hidden_layer_size)].reshape(hidden_layer_size,
                                                                     input_layer_size+1)
    theta2 = nn_params[(input_layer_size+1)*(hidden_layer_size):].reshape(num_labels,
                                                                     hidden_layer_size+1)
    
    theta = (theta1, theta2)
    
    # Implement feed-forward algorithm
    a = x
    for i, layer in enumerate(theta):
        # add ones column
        a = add_ones(a)
        a = sigmoid(mm(a,layer.T))
    
    return np.argmax(a, axis = 1)

nn_predictions = predict(theta, input_layer_size, hidden_layer_size, num_labels, x)
```

```python
def accuracy(p, y):
    return np.mean(nn_predictions==y.flatten())

accuracy(p, y)
```
