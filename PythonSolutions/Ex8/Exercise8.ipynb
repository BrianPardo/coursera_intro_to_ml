## Part 1. Anomaly Detection

In this exercise, we will implement an anomaly detection algorithm to detect anomalous behavior in server computers. The features measure the throughput (mb/s) and latency (ms) of response of each server. While your servers were operating, we collected examples of how they were behaving, and thus have an unlabeled dataset $\{x^{(1)}, \ldots, x^{(m)}\}$. We suspect that the vast majority of these examples are 'normal' (nonanomalous) examples of the servers operating normally, but there might also be some examples of servers acting anomalously within this dataset.

We  will use a Gaussian model to detect anomalous examples in the dataset. First, let's start on a 2D dataset that will allow us to visualize what the algorithm is doing. On that dataset we will fit a Gaussian distribution and then find values that have very low probability and hence can be considered anomalies. After that, we apply the anomaly detection algorithm to a larger dataset with many dimensions.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io
from numpy import matmul as mm
from string import digits
from scipy.optimize import minimize
```

```python
my_data1 = scipy.io.loadmat('ex8data1.mat')
x = my_data1['X']
x_val = my_data1['Xval']
y_val = my_data1['yval']

fig, ax = plt.subplots()
ax.plot(x[:,0],x[:,1],'bx', markersize=3)
ax.set_xlabel('Latency (ms)')
ax.set_ylabel('Throughput (mb/s)');
```

```python
def estimate_gaussian(x):
    m, _ = np.shape(x)
    mu = np.mean(x, axis = 0)
    sigma2 = np.sum((1/m)*(x - mu)**2, axis = 0)
    return mu, sigma2
```

```python
def multivariate_gaussian(x, mu, sigma2):
    exponent = -(x-mu)**2/(2*sigma2)
    prefactor = (2*np.pi*sigma2)**(-0.5)
    return np.product(prefactor*np.exp(exponent), axis=1)
```

```python
mu, sigma2 = estimate_gaussian(x)
p = multivariate_gaussian(x, mu, sigma2)

delta = 0.25
xmesh = np.arange(0, 30, delta).reshape(-1,1)
X1, X2 = np.meshgrid(xmesh, xmesh)
z = np.zeros(np.shape(X1))
for i in range(len(X1)):
    for j in range(len(X1[0])):
        Xtemp = np.array([[X1[i,j],X2[i,j]]])
        z[i,j] = multivariate_gaussian(Xtemp, mu, sigma2)

ax.contour(X1, X2, z, [1e-20, 1e-15, 1e-10, 1e-5, 0.01, 0.1])
fig
```

```python
def select_threshold(y_val, p_val):
    best_epsilon = 0
    best_f1 = 0
    f1 = 0

    delta = (np.max(p_val)-1e-22)/1000
    epsilon = np.arange(1e-22, np.max(p_val), delta)
    for e in epsilon:
        fp = np.sum((p_val < e) & (y_val==0))
        fn = np.sum((p_val > e) & (y_val==1))
        tp = np.sum((p_val < e) & (y_val==1))
    
        prec = tp/(tp+fp);
        rec  = tp/(tp+fn);
        f1 = 2*prec*rec/(prec+rec)
        
        if f1 > best_f1:
            best_f1 = f1
            best_epsilon = e
    return best_epsilon, best_f1
```

```python
epsilon, _ = select_threshold(y_val.flatten(), p)
```

```python
ax.scatter(x[p<epsilon, 0], x[p<epsilon,1], facecolors='none', edgecolors='r')
fig
```

```python
params = scipy.io.loadmat('ex8_movieParams.mat')
params.keys()
x = params['X']
theta = params['Theta']
num_users = params['num_users']
num_movies = params['num_movies']
num_features = params['num_features']
```

```python
def cofi_cost_function(params, y, r, num_users, num_movies, num_features, lambda_r):
    
    x = params[:num_movies*num_features].reshape(num_movies, num_features)
    theta = params[num_movies*num_features:].reshape(num_users, num_features)
    
    J = (1/2)*np.sum(((mm(x,theta.T)-y)*r)**2)
    J += (lambda_r/2)*(np.sum(theta**2)+np.sum(x**2))
    
    x_grad = mm((mm(x,theta.T)-y)*r, theta)+lambda_r*x
    theta_grad = mm(((mm(x,theta.T)-y)*r).T, x) + lambda_r*theta

    grad = np.append(x_grad, theta_grad)
    
    return J, grad
```

```python
num_users = 4
num_movies = 5 
num_features = 3
x = x[:num_movies, :num_features]
theta = theta[:num_users, :num_features]
y = y[:num_movies, :num_users]
r = r[:num_movies, :num_users]

#  Evaluate cost function
J, grad = cofi_cost_function(np.append(x, theta), y, r, num_users, num_movies,num_features, 0);
print(f'Cost at loaded parameters: {J} ')
```

```python
def check_cost_function(lambda_r = 0):
    '''
    Creates a collaborative filering problem 
    to check our cost function and gradients, and outputs the 
    analytical gradients produced by our cost function and the numerical gradients. 
    These two gradient computations should result in very similar values.
    '''
    #Create small problem to efficiently check gradients
    x_t = np.random.rand(4, 3)
    theta_t = np.random.rand(5, 3)

    # Remove most entries to make a sparse matrix
    y = mm(x_t, theta_t.T)
    m, n = np.shape(y)
    y[np.random.rand(m, n) > 0.5] = 0
    r = np.zeros((m, n));
    r[y != 0] = 1

    # Run Gradient Checking
    x = np.random.randn(*np.shape(x_t))
    theta = np.random.randn(*np.shape(theta_t))
    num_users = np.shape(y)[1]
    num_movies = np.shape(y)[0]
    num_features = np.shape(theta_t)[1]
    
    # Join parameters into long array
    params = np.append(x, theta)
    
    J = lambda params: cofi_cost_function(params, y, r,
                                          num_users,
                                          num_movies,
                                          num_features,
                                          lambda_r)[0]
    
    # Compute numerical gradient
    numgrad = compute_numerical_gradient(J, params)
    
    # Compute analytic gradient
    _, grad = cofi_cost_function(np.append(x,theta), y, r, num_users,
                                    num_movies, num_features, lambda_r)
    
    # Print comparison of numerical and analytic gradients
    print('  Gradient    Numerical   Difference')
    print('======================================')
    for i in range(10):
        print(f'{grad[i]:10.5f}  |{numgrad[i]:10.5f} | {grad[i]-numgrad[i]:10.4}')
        
    diff = np.linalg.norm(numgrad-grad)/np.linalg.norm(numgrad+grad)
    print(f'======================================\nRelative Difference: {diff:10.5}')
```

```python
def compute_numerical_gradient(J, theta):
    """This function takes a cost function and theta and returns the numerical gradient"""
    numgrad = np.zeros(np.shape(theta));
    perturb = np.zeros(np.shape(theta));
    e = 1e-4;
    for p in range(len(theta)):
        # set perturbation vector
        perturb[p] = e;
        loss1 = J(theta - perturb);
        loss2 = J(theta + perturb);
        
        # Compute Numerical Gradient
        numgrad[p] = (loss2-loss1)/(2*e);
        perturb[p] = 0
        
    return numgrad
```

```python
def load_movie_list(file_name):
    movies = {}
    with open(file_name, encoding='latin-1') as movie_list:
        for i, movie in enumerate(movie_list):
            movies[i] = movie.lstrip(digits).strip()
    return movies
```

```python
# Load movie list
movie_list = load_movie_list('movie_ids.txt')

# Initialize my ratings
my_ratings = np.zeros((1682, 1))

# Check the file movie_idx.txt for id of each movie in our dataset
# For example, Toy Story (1995) has ID 1, so to rate it "4", we set
my_ratings[0] = 4
# Or suppose we did not enjoy Silence of the Lambs (1991), we set
my_ratings[97] = 2

# Let's select a few movies we liked/did not like and give ratings as follows:
my_ratings[6] = 3
my_ratings[11]= 5
my_ratings[53] = 4
my_ratings[63]= 5
my_ratings[65]= 3
my_ratings[68] = 5
my_ratings[182] = 4
my_ratings[225] = 5
my_ratings[354]= 5

print('New user ratings:')
print('=============================')
for i in range(len(my_ratings)):
    if my_ratings[i] > 0:
        print(f'Rated {my_ratings[i][0]:.0f} for {movie_list[i]}')
```

```python
def normalize_ratings(y, r):
    y_mean = np.sum(y, axis=1)/np.sum(r, axis=1)
    y_norm = (y-y_mean.reshape(-1,1))*r
    return y_norm, y_mean
```

```python
y_norm, y_mean = normalize_ratings(y, r)
```

```python
y = movies['Y']
r = movies['R']

# Y is a 1682x943 matrix, containing ratings (1-5) of 1682 movies by 943 users
# R is a 1682x943 matrix, where R(i,j) = 1 if and only if user j gave a rating to movie i
# Add our own ratings to the data matrix
y = np.append(my_ratings, y, axis=1)
r = np.append(my_ratings != 0, r, axis=1)

#  Normalize Ratings
y_norm, y_mean = normalize_ratings(y, r)

#  Useful Values
num_movies, num_users = np.shape(y)
num_features = 10
lambda_r = 10

# Set Initial Parameters (Theta, X)
x = np.random.randn(num_movies, num_features)
theta = np.random.randn(num_users, num_features)
initial_params = np.append(x, theta)

# Set options for fmincg
options = {'jac':True, 'method':'TNC', 'options': {'maxiter':150}}

args = (y_norm, r, num_users, num_movies, num_features, lambda_r)
params = np.append(x, theta)
optim = minimize(cofi_cost_function, params, args=args, **options)

params = optim.x
cost = optim.fun
```

```python
x_opt = params[:num_movies*num_features].reshape(num_movies, num_features)
theta_opt = params[num_movies*num_features:].reshape(num_users, num_features)
```

```python
p = mm(x_opt, theta_opt.T)
my_p = (p + y_mean.reshape(-1,1))
top10 = np.argsort(my_p[:,0])[-10:][::-1]

print('Top 10 predicted ratings:')
print('========================================')
for movie in top10:
    print(f'Predict {my_p[movie, 0]:.1f} for {movie_list[movie]}')
```
