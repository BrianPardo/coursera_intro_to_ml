## Part 1. Regularized Linear Regression
In the first half of the exercise, we will implement regularized linear regression to predict the amount of water flowing out of a dam using the change of water level in a reservoir. In the next half, we will go through some diagnostics of debugging learning algorithms and examine the effects of bias vs. variance.

```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize
from numpy import matmul as mm

mydata = scipy.io.loadmat('ex5data1.mat')
```

```python
x = mydata['X']
y = mydata['y']
x_val = mydata['Xval']
y_val = mydata['yval']
x_test = mydata['Xtest']
y_test = mydata['ytest']
```

```python
fig1, ax = plt.subplots()
ax.plot(x.flatten(), y.flatten(), 'x')
ax.set_title('Historical record of water flow')
ax.set_xlabel('Change in water level (x)')
ax.set_ylabel('Water flowing out of the dam (y)');
```

```python
def add_ones(x):
    # add ones column to x
    ones = np.ones((len(x),1))
    return np.append(ones,x,axis=1)
```

```python
def lr_cost_function(theta, x, y, lambda_r):
    m = len(x)
    
    x = add_ones(x)
    theta_reg = np.append(np.array([0]), theta[1:]).reshape(-1,1)
    theta = theta.reshape(-1,1)
    
    h = mm(x, theta)
    
    # Compute the cost and gradient without regularization
    J = (1/(2*m))*mm((h-y).T,h-y)
    grad = (1/m)*mm(x.T, h-y)
    
    # Regularize cost and gradient
    
    J = J + (lambda_r/(2*m))*np.sum(theta_reg**2)
    grad = grad + (lambda_r/m)*theta_reg
    
    return float(J), grad.flatten()
```

```python
initial_theta = np.array([1,1])
lambda_r = 1
lr_cost_function(initial_theta, x, y, lambda_r)
```

```python
# Learn the parameters for our model
initial_theta = np.array([1,1])
lambda_r = 100
inputs = (x, y, lambda_r)
kwargs = {'jac':True, 'method':'TNC', 'options': {'maxiter':50}}

optim = minimize(lr_cost_function, initial_theta, inputs, **kwargs)
theta_opt = optim.x
```

```python
# Plot the linear regression model with data
ax.plot(x, mm(add_ones(x),theta_opt),'y')
fig1
```

```python
# Let's write a function to compute the cost function as we add examples
def learning_curve(x, y, x_val, y_val, lambda_r):
    (m, n) = np.shape(x)
    
    error_train = np.zeros(m)
    error_val   = np.zeros(m)
    
    initial_theta = np.ones((n+1,))
    
    kwargs = {'jac':True, 'method':'TNC', 'options': {'maxiter':50}}
    
    for i in range(0,m):
        inputs = (x[:i+1,:], y[:i+1,:], lambda_r)
        optim = minimize(lr_cost_function, initial_theta, inputs, **kwargs)
        initial_theta = optim.x
        error_train[i], _ = lr_cost_function(initial_theta, *inputs)
        error_val[i],   _ = lr_cost_function(initial_theta, x_val, y_val, lambda_r)
    
    return error_train, error_val 
```

```python
error_train, error_val = learning_curve(x, y, x_val, y_val, 0)
error_val
```

```python
fig2, ax = plt.subplots()

ax.plot(error_train, label = 'Training error')
ax.plot(error_val, label = 'Validation error')
ax.legend(bbox_to_anchor=(1,1), loc='upper left');
ax.set_xlabel('Training examples')
ax.set_ylabel('Cost');
```

```python
def poly_features(arr, p):
    '''Returns array up to x**p'''
    for n in range(2,p+1):
        arr = np.append(arr, arr[:,0:1]**n, axis=1)
    return arr

p = 8
x_poly = poly_features(x, p)
```

```python
def normalize_features(arr, arr_mean, arr_std): 
    return (arr-arr_mean)/arr_std

x_poly_norm = normalize_features(x_poly, np.mean(x_poly,axis=0),np.std(x_poly, axis=0))
```

```python
# Train the model
initial_theta = np.zeros(len(x_poly_norm[0])+1)
inputs = (x_poly_norm, y, lambda_r)
optim = minimize(lr_cost_function, initial_theta, inputs, **kwargs)
```

```python
theta = optim.x
```

```python
fig3, ax = plt.subplots()
ax.plot(x, y, 'x')
delta = 0.1
xp = np.arange(np.min(x)-15, np.max(x)+15, delta).reshape(-1,1)
xp_poly = poly_features(xp, p)
xp_poly_norm = normalize_features(xp_poly, np.mean(x_poly, axis=0), np.std(x_poly, axis=0))
yp = mm(add_ones(xp_poly_norm), theta)
ax.plot(xp, yp);
```

```python
x_val_poly = poly_features(x_val, p)
x_val_poly_norm = normalize_features(x_val_poly, np.mean(x_poly, axis=0), np.std(x_poly, axis=0))
error_train, error_val = learning_curve(x_poly_norm, y, 
                                        x_val_poly_norm, y_val, lambda_r)

fig4, ax = plt.subplots()
ax.plot(error_train, label = 'Training error')
ax.plot(error_val, label = 'Validation error')
ax.legend(bbox_to_anchor=(1,1), loc='upper left');
ax.set_xlabel('Training examples')
ax.set_ylabel('Cost');
```
