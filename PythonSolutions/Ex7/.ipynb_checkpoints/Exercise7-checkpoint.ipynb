We will implement the $K$-means algorithm and use it for image compression. We first start on an example 2D dataset that will help gain an intuition of how the algorithm works. ### Implementing $K$-means
The $K$-means algorithm is a method to automatically cluster similar data examples together. Concretely, you are given a training set and want to group the data into a few cohesive 'clusters'. The intuition behind $K$-means is an iterative procedure that starts by guessing the initial centroids, and then refines this guess by repeatedly assigning examples to their closest centroids and then recomputing the centroids based on the assignments.
    The $K$-means algorithm is as follows:

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
from numpy import matmul as mm
from mpl_toolkits.mplot3d import Axes3D
import utils
```

```python
def read_image(image_file):
    im = plt.imread(image_file)
    imshape = np.shape(im)
    return im.reshape(imshape[0]*imshape[1],imshape[2])

image = read_image('bird_small.png')
```

```python
k = 16
initial_centroids = utils.initialize_centroids(image, k)
idx, centroids = utils.run_kmeans(image, k, init_centroids=initial_centroids, iters = 10)
```

```python
image_compress = centroids[idx]
(m, n) = np.shape(image_compress)
w = np.sqrt(m).astype('int64')
image_compress = image_compress.reshape((w,w,n))
image_original = image.reshape((w,w,n))

fig1, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(image_compress)
ax2.imshow(image_original);
ax1.axis('off')
ax2.axis('off');
```

```python
fig2 = plt.figure()
ax = fig2.add_subplot(projection='3d')
for color in range(k):
    im_color = image[idx==color]
    rgb = (centroids[color][0],centroids[color][1],centroids[color][2])
    ax.scatter(im_color[:,0],im_color[:,1],im_color[:,2], color = rgb);
```

```python
my_data1 = scipy.io.loadmat('ex7data1.mat')
x = my_data1['X']

k=5
initial_centroids = utils.initialize_centroids(x, k)
idx, centroids = utils.run_kmeans(x, k, init_centroids=initial_centroids, iters = 5)

fig3 = plt.figure()
ax = fig3.add_subplot()
for c in range(k):
    cluster = x[idx==c]
    ax.plot(cluster[:,0], cluster[:,1], 'x')
```

```python
my_data2 = scipy.io.loadmat('ex7data2.mat')
x = my_data2['X']

k=3
iters = 10
centroid_history = np.zeros((iters+1, k, 2))
centroids = utils.initialize_centroids(x, k)
centroid_history[0,:,:] = centroids
for i in range(iters):
    idx, centroids = utils.run_kmeans(x, k, init_centroids=centroids, iters = 1)
    centroid_history[i+1,:,:] = centroids

centroid_history[:,0,:]
fig4 = plt.figure()
ax = fig4.add_subplot()
for c in range(k):
    cluster = x[idx==c]
    ax.plot(cluster[:,0], cluster[:,1], 'x')
    ax.plot(centroid_history[:,c,0],centroid_history[:,c,1],'o-')
```

## Part 2. Principle Component Analysis

In this exercise, we will use principal component analysis (PCA) to perform dimensionalty reduction. Let's first experiment with an example 2D dataset to get intuition on how PCA works, and then use it on a bigger dataset of 5000 face images.

```python
# Load the sample dataset for PCA

my_data1 = scipy.io.loadmat('ex7data1.mat')
x = my_data1['X']
```

```python
fig5 = plt.figure(figsize=(4,4))
ax = fig5.add_subplot()
ax.plot(x[:,0], x[:,1], 'x')
ax.set_xbound(0.5,6.5)
ax.set_ybound(2,8)
```

```python
x_norm, mu, std = utils.normalize(x)

U, V = utils.pca(x_norm)
arr = np.arange(2).reshape(-1,1)
```

```python
vec1 = mu + 1.5*mm(arr,U[0:1])*V[0]*std
vec2 = mu + 1.5*mm(arr,U[1:])*V[1]*std
```

```python
ax.plot(vec1[:,0],vec1[:,1],'-')
ax.plot(vec2[:,0],vec2[:,1],'-')
fig5
```

```python
z = utils.project_data(x_norm, U, 1)
x_proj = utils.recover_data(z, U, 1)
```

## Part 3. PCA for data visualization

```python
image = read_image('bird_small.png')

k = 16
initial_centroids = utils.initialize_centroids(image, k)
idx, centroids = utils.run_kmeans(image, k, init_centroids=initial_centroids, iters = 10)

image_compress = centroids[idx]
(m, n) = np.shape(image_compress)
w = np.sqrt(m).astype('int64')
image_compress = image_compress.reshape((w,w,n))
image_original = image.reshape((w,w,n))

fig1, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(image_compress)
ax2.imshow(image_original);
ax1.axis('off')
ax2.axis('off');

fig2 = plt.figure()
ax = fig2.add_subplot(projection='3d')
for color in range(k):
    im_color = image[idx==color]
    rgb = (centroids[color][0],centroids[color][1],centroids[color][2])
    ax.scatter(im_color[:,0],im_color[:,1],im_color[:,2], color = rgb, marker = '.');
```

```python
im_norm, mu, std = utils.normalize(image)
U, S = utils.pca(im_norm)
im_proj = utils.project_data(im_norm, U, 2)
fig = plt.figure()
ax = fig.add_subplot()
for color in range(k):
    im_color = im_proj[idx==color]
    rgb = (centroids[color][0],centroids[color][1],centroids[color][2])
    ax.scatter(im_color[:,0],im_color[:,1], color = rgb, marker = '.')
```
